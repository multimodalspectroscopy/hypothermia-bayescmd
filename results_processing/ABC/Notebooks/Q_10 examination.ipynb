```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
temperatures = np.linspace(-10,0,500)
Q10_range = np.linspace(0.01,1,100)
```

```python
def rate(T_2,T_1, Q10):

    return Q10**((T_2-T_1)/10)

rate_vec = np.vectorize(rate)
```

```python
sns.set_style("whitegrid")
fig, ax = plt.subplots(1, figsize=(8,6))
temperatures = np.linspace(-15,15,500)
Q10_range = [3,2,1,0.5]
R2 = np.zeros((len(Q10_range), 500))
for ii, Q10 in enumerate(Q10_range):
    line = plt.plot(temperatures, rate_vec(temperatures,0,Q10), label=Q10)
    plt.plot(10,rate(10,0,Q10), 'o', color=line[0].get_color())
    plt.plot(-10,rate(-10,0,Q10), 'x', color=line[0].get_color())

ax.set_xlabel('Difference in temperature ($T_{2}-T_{1}$) [$^{\circ}C$]', size=14)    
ax.set_ylabel('Ratio of reaction rates ($R_{2}/R_{1}$)', size=14)
plt.legend(title="$Q_{10}$", fontsize=12, title_fontsize=14)
plt.savefig('/home/buck06191/Dropbox/phd/Final-Thesis/Figures/Hypothermia/Q10TemperatureCoefficient.png', dpi=300)
```
